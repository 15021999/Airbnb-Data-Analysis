```python
# let us check all the null values in our dataset
null_values_in_data=air_bnb.isnull().sum().sort_values(ascending=False)
print(null_values_in_data)
```

```python
air_bnb_copy=air_bnb.copy()
```

```python
air_bnb_copy.head()
```

```python
# Drop columns: 'reviews_per_month' and 'last_review'
air_bnb_copy.drop(['reviews_per_month','last_review'],axis=1,inplace=True)
```

```python
# replace null value from host_name column to NOT AVAILABLE

list_of_index_host_name_nullvalues=list(air_bnb_copy[pd.isnull(air_bnb_copy["host_name"])].index.values)
air_bnb_copy.iloc[list_of_index_host_name_nullvalues,[3]] ='NOT AVAILABLE'
```

```python
#also drop name colume because it is not holding any intact information
air_bnb_copy.drop(['name'],axis=1,inplace=True)
```

```python
# check our data set if it is clear or not
null_values_in_data=air_bnb_copy.isnull().sum().sort_values(ascending=False)
print(null_values_in_data)
```

```python
air_bnb_copy.describe()
```

```python
# so let's check where price is 0
air_bnb_copy[air_bnb_copy['price']==0]
```

```python
mean_value_of_price = air_bnb_copy['price'].mean()
```
